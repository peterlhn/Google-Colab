**Toss a Coin and see "Head or Tail"**

```
import numpy as np
np.random.seed(123)

# Randomly generate 0 or 1
coin = np.random.randint(0,2)     

if coin == 0:
    print("head")
else:
    print("tails")
```

**Roll the dice. Use randint() to create the variable dice. (MOVE ONE each time)**


* If dice is 1 or 2, you go one step down for 1.
* if dice is 3, 4 or 5, you go one step up for 1.
* Else, you throw the dice again. The number of eyes is the number of steps you go up.

```
import numpy as np
np.random.seed(123)

# Starting step
step = 50

# Roll the dice
dice = np.random.randint(1,7)

# Finish the control construct
if dice <= 2 :
    step = step - 1
elif dice > 2 and dice < 6 :
    step = step + 1
else :
    step = step + np.random.randint(1,7)

# Print out dice and step
print(dice)
print(step)
```

##**Random Walk**

```
import numpy as np
np.random.seed(123)
outcome = []
tails = [0]

# Toss the coin 10 times, and push the result to a list 
for x in range(10):
    coin = np.random.randint(0, 2)

    # Push the result to the outcome list 
    if coin == 0:
        outcome.append("heads")
    else: 
        outcome.append("tails")

    # Track how many "Tail" appear in this round
    tails.append(tails[x] + coin )    # Add 1 if coin flip a Tail 

# Print out the result
print(outcome)

# Find out how many time appeared by the end of 10 times
print(tails)   
print(max(tails))



```

**Roll the dice. Use randint() to create the variable dice. (RANDOM WALK)**


* If dice is 1 or 2, you go one step down for 1.
* if dice is 3, 4 or 5, you go one step up for 1.
* Else, you throw the dice again. The number of eyes is the number of steps you go up.

```
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)
random_walk = [0]

# Roll the dice 100 times, and push the result to a list 

for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]
    
    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:                 # If random_walk start at 0, step - 1 will not make sense
        step = max(0, step - 1)   # make sure that a "step" never goes below 0
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)


# ===== Visualize random_walk by matplotlib ====== #

# Plot random_walk
plt.plot(random_walk)
plt.show()

```

## **Multiple Walks**

**Repect the "Random Walk" 10 times**

```
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)
random_walk = [0]

# Initialize all_walks to contain all "Random Walk"
all_walks = []

# Simulate "random walk" 100 times
for i in range(100) :

    # Random Walk - ONE walk - with 100 step
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
     

        random_walk.append(step)

    # Append "random_walk" x 10 to "all_walks" <<- a'LIST'
    all_walks.append(random_walk)

# ===== Visualize random_walk by matplotlib ====== #

# Convert all_walks to Numpy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show - FAILED due to datastructure
# plt.plot(np_aw)
# plt.show()
# plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()

# === Distribution === #
# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()

# Find the Odd if over 60 points by the end of each walk
print(str(np.mean(ends > 60)) + ' % walk is over 60 pt')
```
